```python
#import library
import networkx as nx
import matplotlib.pyplot as plt
```

# Buatlah visualisasi jaringan yang merepresentasikan jalur transportasi yang menghubungkan kota-kota besar di Jawa Barat.

```python
#Preparing Data
Graf = nx.Graph()
```

## Add Nodes

```python
Graf.add_node("Sukabumi",pos=(2,0))
Graf.add_node("Cianjur",pos=(3,2))
Graf.add_node("Bogor",pos=(1,3))
Graf.add_node("Bekasi",pos=(3,5))
Graf.add_node("Cikampek",pos=(4,4))
Graf.add_node("Subang",pos=(7,3))
Graf.add_node("Purwakarta",pos=(5,3))
Graf.add_node("Bandung",pos=(6,0))
Graf.add_node("Sumedang",pos=(8,1))
Graf.add_node("Indramayu",pos=(10,6))
Graf.add_node("Cirebon",pos=(12,3))
Graf.add_node("Tasikmalaya",pos=(10,-2))
Graf.add_node("Banjar",pos=(13,-3))
```

```python
Graf.nodes()
nx.draw(Graf, with_labels=True)
plt.show()
```

## Add Edges


```python
Graf.add_edge("Bekasi","Cikampek")
Graf.add_edge("Sukabumi","Cianjur")
Graf.add_edge("Bogor","Cianjur")
Graf.add_edge("Bogor","Sukabumi")
Graf.add_edge("Cikampek","Purwakarta")
Graf.add_edge("Cikampek","Subang")
Graf.add_edge("Cikampek","Indramayu")
Graf.add_edge("Purwakarta","Bandung")
Graf.add_edge("Purwakarta","Subang")
Graf.add_edge("Cianjur","Bandung")

Graf.add_edge("Bandung","Sumedang")

Graf.add_edge("Subang","Cirebon")
Graf.add_edge("Indramayu","Cirebon")
Graf.add_edge("Sumedang","Cirebon")
Graf.add_edge("Bandung","Tasikmalaya")
Graf.add_edge("Bandung","Subang")
Graf.add_edge("Tasikmalaya","Banjar")
Graf.add_edge("Tasikmalaya","Sumedang")
```

```python
Graf.edges()
nx.draw(Graf, with_labels=True)
plt.show()
```

## Plot

```python
pos=nx.get_node_attributes(Graf,'pos')
```

```python
nx.draw(Graf,pos, with_labels=True)
plt.show()
```

# Tentukan kota terpenting berdasarkan nilai degree centrality dan betweenness centrality.

```python
# Degree Centrality
nx.degree_centrality(Graf)
```

```python
degree = nx.degree_centrality(Graf)
plt.bar(range(len(degree)), list(degree.values()), align='center')
plt.xticks(range(len(degree)), list(degree.keys()),rotation='vertical')
plt.title("Degree Centrality")


plt.show()
```

```python
#Betweenness Centrality
nx.betweenness_centrality(Graf)
```

```python
between = nx.betweenness_centrality(Graf)
plt.bar(range(len(between)), list(between.values()), align='center')
plt.xticks(range(len(between)), list(between.keys()),rotation='vertical')
plt.title("Betweenness Centrality")


plt.show()
```

Kota terpenting menurut degree dan betweenness centrality adalah Bandung


# Update ukuran node berdasarkan nilai degree centrality

```python
nx.draw(Graf,pos, with_labels=True, node_size=[v * 5000 for v in degree.values()], node_color="tab:purple")
plt.show()
```
